```python
# import the relevant classes
from IWR1443_Config import IWR1443_Config
from IWR1443_Processor import IWR1443_Processor
from IWR1443_Streamer import IWR1443_Streamer
```

## Note on configuring for linux based devices
1. to locate the serial ports, use the commnad: "ls /dev/ttyACM*" to find the serial ports
2. if there is an error saying that access is denied, use the command "sudo usermod -a -G dialout $USER"

```python
config = IWR1443_Config("1443config.cfg",
                        verbose=False,
                        translate_from_JSON= False,
                        enable_serial=True,
                        CLI_port='/dev/ttyACM0',
                        Data_port='/dev/ttyACM1')

processor = IWR1443_Processor(
    config_parameters=config.config_params,
    enable_plotting=True,
    jupyter=True,
    verbose=False)

data_file = "xwr14xx_processed_stream_2023_04_05T13_04_55_356.dat"
streamer = IWR1443_Streamer(
    processor=processor,
    enable_serial=True,
    data_file=data_file,
    CLIPort=config.CLIport,
    DataPort=config.Dataport,
    verbose=False,
    refresh_rate=50.0
)

streamer.stream()
```
